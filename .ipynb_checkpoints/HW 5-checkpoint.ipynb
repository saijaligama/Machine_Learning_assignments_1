```python
df1=pd.read_csv(r'C:\Users\saija\Downloads/Frogs_MFCCs.csv')
df1=df1.iloc[:,22 :-1]
df=pd.read_csv(r'C:\Users\saija\Downloads/Frogs_MFCCs.csv')
df=df.iloc[:, :-4]

cols = ['Family','Genus','Species']
predicted_Labels = 'predicted'
labels = df1
avg_silhouette_scores_k_1 = []
k_list_1 = []
predicted = []
hamming_dist=[]
hamming_dist_1=[]
Most_Commom_ClassInLabel = {}
Common_Class={}
silhouette_avg_Scores = []
for k in range(2,5):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    pred=kmeans.predict(df)    
    silhouette_avg_Scores.append(silhouette_score(df, pred))
    predicted.append(pred)
best_k = silhouette_avg_Scores.index(max(silhouette_avg_Scores)) + 2
k_list_1.append(best_k)
avg_silhouette_scores_k_1.append(max(silhouette_avg_Scores))
predicted[np.argmax(silhouette_avg_Scores)]
labels[predicted_Labels] = predicted[np.argmax(silhouette_avg_Scores)]
for each_cluster in range(best_k):
    Each_Cluster_Data = labels[labels[predicted_Labels]==each_cluster]
    for x in cols:
         Most_Commom_ClassInLabel[x] = Counter(Each_Cluster_Data[x]).most_common(1)[0][0]
    Common_Class[each_cluster] = Most_Commom_ClassInLabel
print("The mijority class label in the cluster {} is {}".format(best_k, Common_Class))
```

```python
parameters = {"C": np.logspace(-2, 5, 10), 'gamma': [i for i in np.arange(0.1,2,0.1)]}
cols = ['Family','Genus','Species']

for x in cols:
    print(x)
    Y_train=Y_Train[x]
    print('1')
    Y_test=Y_Test[x]
    svc = SVC(kernel="rbf", decision_function_shape='ovr')
    print('1')
    gridSearchCV=GridSearchCV(svc, parameters, cv=10)
    print('1')
    gridSearchCV.fit(X_Train, Y_train)
    print('1')
    bestestimators = gridSearchCV.best_estimator_
    print('1')
    y_predict = bestestimators.predict(X_Test)
    print(x)
    print("Using RawData, The best params are: ", gridSearchCV.best_params_)    
    print("Using RawData, The accuracy score is: ", accuracy_score(Y_test, y_predict))
    print("Using RawData, The hamming loss is: ", hamming_loss(Y_test, y_predict))
    
```

```python
parameters = {"C": [1,10,100,1000]}
cols = ['Family','Genus','Species']

for x in cols:
    print(x)
    Y_train=Y_Train[x]
    print('1')
    Y_test=Y_Test[x]    
    scaler = StandardScaler()
    LassoSD=scaler.fit(X_Train)
    LassoTR= scaler.transform(X_Train)
    LassoTR_Test= scaler.transform(X_Test)
    SVCLinear=LinearSVC(penalty="l1", multi_class='ovr', dual=False,max_iter = 2000)
    gridCV = GridSearchCV(SVCLinear, parameters, cv=10, n_jobs=-1)
    gridCV.fit(LassoTR, Y_train)
    bestClf = gridCV.best_estimator_
    y_predict = bestClf.predict(LassoTR_Test)
    print(x)
    print("Using L1 penalized SVM, The best params are: ", gridCV.best_params_)
    print("Using L1 penalized SVM, The accuracy score is: ", accuracy_score(Y_test, y_predict))
    print("Using L1 penalized SVM, The hamming loss is: ", hamming_loss(Y_test, y_predict))
    EM=np.all(y_predict == Y_test).mean()
    print("Exact Match", EM)



```

```python
from imblearn.over_sampling import SMOTE
parameters = {"C": [1,10,100,1000]}
cols = ['Family','Genus','Species']

for x in cols:
    print(x)
    Y_train=Y_Train[x]
    Y_test=Y_Test[x] 
    oversample = SMOTE()
    X, y = oversample.fit_resample(X_Train, Y_train)    
    scaler = StandardScaler()
    LassoSD=scaler.fit(X)
    LassoTR= scaler.transform(X)
    LassoTR_Test= scaler.transform(X_Test)
    SVCLinear=LinearSVC(penalty="l1", multi_class='ovr', dual=False,max_iter = 2000)
    gridCV = GridSearchCV(SVCLinear, parameters, cv=10, n_jobs=-1)
    gridCV.fit(LassoTR, y)
    bestClf = gridCV.best_estimator_
    y_predict = bestClf.predict(LassoTR_Test)
    print(x)
    print("Using L1 penalized SVM, The best params are: ", gridCV.best_params_)
    print("Using L1 penalized SVM, The accuracy score is: ", accuracy_score(Y_test, y_predict))
    print("Using L1 penalized SVM, The hamming loss is: ", hamming_loss(Y_test, y_predict))
    EM=np.all(y_predict == Y_test).mean()
    print("Exact Match", EM)

    
```
